#Introducing Artificial Neural Networks

Redes neuronales
- x:  entradas de 1 hasta n. Resulta que cada nodo de entrada está asociado a un valor numérico, que puede ser cualquier número real.
- W: pesos de la entradas
- Suma ponderada de pesos
- f:  dunción de activación
- Conexiones. Del mismo modo, cada conexión que parte del nodo de entrada tiene asociado un peso, que también puede ser cualquier número real.

 $$ y = f (∑ _{i=1}^{D} w_i*x_i )$$  
  o
$$y = f(w_1*x_1 + w_2*x_2 + ... w_D*x_D)$$

Este resultado será la entrada para una función de transferencia o activación.



$f(x)= 0$ si $x<0 $   
$f(x) = 0,5 $ si $x=0$  
$f(x) = 1 $ si $x>0$

Por último, el perceptrón puede tener un parámetro adicional, llamado sesgo, que puedes considerar como el peso asociado a un nodo de entrada adicional que se fija permanentemente en 1.

##Perceptrones multicapa

Las redes de perceptrones son perceptrones multicapa, ambién se conocen como "redes neuronales feed-forward".    

Las capas actúan de forma muy parecida a las neuronas biológicas sobre las que has leído más arriba: las salidas de una capa sirven como entradas para la capa siguiente.  

Entre las capas, puedes distinguir una capa de entrada, capas ocultas y una capa de salida. Los perceptrones multicapa suelen estar totalmente conectados.
